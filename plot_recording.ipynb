```python
import mne
import matplotlib.pyplot as plt
```

```python
# Path to the .fif file
file_path = 'recordings/recording_8077.raw.fif'
```

```python
# Read the raw data
raw = mne.io.read_raw_fif(file_path, preload=True)
```

```python
# Print raw data info
raw.info
```

```python
# Observe annotations
annotations = raw.annotations
annotations.description, annotations.onset
```

```python
# Plot the raw data (only the first 10 seconds)
raw.plot()
plt.show()
```
