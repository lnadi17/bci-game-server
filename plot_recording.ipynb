```python
import mne
import matplotlib.pyplot as plt
```

```python
# Path to the .fif file
file_path = 'recordings/recording.raw.fif'
```

```python
# Read the raw data
raw = mne.io.read_raw_fif(file_path, preload=True)
```

```python
# Print raw data info
raw.info
```

```python
# Plot the raw data (only the first 10 seconds)
raw.crop(tmin=0, tmax=10).plot()
plt.show()
```
